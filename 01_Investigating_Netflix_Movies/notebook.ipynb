## The data
### **netflix_data.csv**
| Column | Description |
|--------|-------------|
| `show_id` | The ID of the show |
| `type` | Type of show |
| `title` | Title of the show |
| `director` | Director of the show |
| `cast` | Cast of the show |
| `country` | Country of origin |
| `date_added` | Date added to Netflix |
| `release_year` | Year of Netflix release |
| `duration` | Duration of the show in minutes |
| `description` | Description of the show |
| `genre` | Show genre |

```python
# Importing pandas and matplotlib
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Read in the Netflix CSV as a DataFrame
netflix_df = pd.read_csv("netflix_data.csv")
```

```python
# Filter the DataFrame to include only rows where the 'type' is 'Movie'
type_df = netflix_df[(netflix_df['type'] == 'Movie')]

# Further filter the DataFrame to include only movies released between 1990 and 1999
date_df = type_df[(type_df['release_year'] <= 1999) & (type_df['release_year'] >= 1990)]

# Create a histogram to visualize the distribution of movie durations for the selected years
plt.hist(date_df['duration'], bins=20)
plt.xlabel('Duration (minutes)')  # Label for the x-axis
plt.ylabel('Frequency')            # Label for the y-axis
plt.title('Movie Duration Distribution (1990s)')  # Title of the histogram
plt.show()  # Display the histogram

# Calculate and print the most common movie duration in the filtered dataset
duration = date_df['duration'].mode()[0]
print(duration)

# Filter the DataFrame to include only 'Action' genre movies from the 1990s
genre_df = date_df[date_df['genre'] == 'Action']

# Initialize a counter for movies with a duration of less than 90 minutes
short_movie_count = 0

# Iterate through the rows of the Action genre DataFrame to count short movies
for label, row in genre_df.iterrows():
    if row['duration'] < 90:
        short_movie_count += 1  # Increment the counter for each short movie

# Print the total count of short Action movies
print(short_movie_count)
```
